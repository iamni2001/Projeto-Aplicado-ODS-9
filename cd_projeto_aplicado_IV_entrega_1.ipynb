# **Introdução**

O desenvolvimento industrial, aliado à inovação tecnológica e à melhoria
da infraestrutura, desempenha um papel fundamental no crescimento
econômico e social de um país. Nesse contexto, o Objetivo de
Desenvolvimento Sustentável 9 (ODS 9), proposto pela Organização das
Nações Unidas, busca promover a construção de infraestruturas
resilientes, incentivar a industrialização inclusiva e sustentável, além de
fomentar a inovação.


De acordo com a Organização das Nações Unidas (2023), ainda existem
grandes desigualdades no acesso à infraestrutura e à tecnologia entre
diferentes regiões do mundo, o que impacta diretamente o
desenvolvimento econômico e a qualidade de vida da população. Além
disso, países em desenvolvimento enfrentam desafios significativos
relacionados à modernização industrial e à adoção de soluções
tecnológicas inovadoras.


Diante desse cenário, a análise de dados relacionados à indústria,
inovação e infraestrutura torna-se essencial para compreender padrões,
identificar problemas e propor melhorias. Assim, este projeto tem como
foco explorar dados relacionados ao ODS 9, buscando extrair informações
relevantes que possam contribuir para uma melhor compreensão do tema.



# **Motivações e justificativa**

##Motivações

• Transição Energética: A indústria é um dos setores mais difíceis de descarbonizar (hard-to-abate); entender seus padrões de emissão é o primeiro passo para a inovação.

• Pressão de Mercado: Investidores e consumidores exigem cada vez mais transparência ambiental (critérios ESG) das indústrias.

• Adoção de Novas Tecnologias: Verificar se a implementação de energias renováveis e eficiência industrial está refletida nos dados ao longo dos anos.

##Justificativa

O setor industrial possui uma dinâmica de emissões fortemente atrelada a ciclos econômicos e avanços tecnológicos, o que gera dados com tendências claras e, em alguns casos, sazonalidades ligadas à produção de commodities. Analisar essas emissões como uma série temporal justifica-se pela necessidade de identificar "pontos de inflexão" (onde as emissões começam a cair apesar do crescimento da produção) e para prever se as metas de "Net Zero" são estatisticamente viáveis dentro do cronograma atual.

# **Descrição da base de dados**

A base de dados será extraída do repositório oficial do Our World in Data (OWID), que compila dados do Global Carbon Project.

• Conteúdo: Séries históricas de emissões de (medidas em milhões de toneladas).

• Granularidade Temporal: Dados anuais, geralmente cobrindo desde meados do século XVIII até o ano mais recente disponível.

- Variáveis Principais:

  • Entity: País ou região.

  • Year: Ano da observação (eixo temporal).

  • Annual CO2 emissions: Total de emissões.

  • Industry/Manufacturing emissions: Recorte específico para o setor industrial (quando disponível no dataset expandido).

• Formato: Arquivo .csv estruturado de forma "longa" (tidy data), facilitando a manipulação em ferramentas como Python (Pandas) ou R.

• Fontes dos dados: Global Carbon Budget (2025); Dados populacionais baseados em várias fontes (2024).

##Declaração oficial do site Our World in Data (OWD)

“As emissões humanas de gases de efeito estufa são o principal fator das mudanças climáticas na atualidade. O CO2 e outros gases de efeito estufa, como o metano e o óxido nitroso, são emitidos quando queimamos combustíveis fósseis, produzimos materiais como aço, cimento e plásticos e cultivamos os alimentos que consumimos. Se quisermos reduzir essas emissões, precisamos transformar nossos sistemas energéticos, indústrias e sistemas alimentares.

Ao mesmo tempo, precisamos combater a pobreza energética, os baixos padrões de vida e a má nutrição, que continuam sendo problemas enormes para bilhões de pessoas.

Os avanços tecnológicos podem nos permitir fazer ambas as coisas. Os preços da energia solar, eólica e das baterias despencaram nas últimas décadas, tornando-se cada vez mais competitivos em relação aos custos das alternativas aos combustíveis fósseis. Mais progressos podem nos permitir fornecer energia limpa e barata para todos. Mudanças políticas são essenciais para criar um sistema que apoie a rápida descarbonização.

As emissões ainda estão aumentando em muitas partes do mundo. No entanto, vários países conseguiram reduzir suas emissões nas últimas décadas. Com tecnologias acessíveis e de baixo carbono, outros países podem aumentar seus padrões de vida sem seguir o caminho de altas emissões de carbono trilhado pelos países ricos no passado.”

# **Introdução**

O desenvolvimento industrial, aliado à inovação tecnológica e à infraestrutura, desempenha papel central no crescimento econômico, mas também está diretamente relacionado ao aumento das emissões de gases de efeito estufa. Nesse contexto, os Objetivos de Desenvolvimento Sustentável, especialmente o ODS 9 (indústria, inovação e infraestrutura) e o ODS 13 (ação contra a mudança global do clima), propostos pela Organização das Nações Unidas, destacam a necessidade de promover o desenvolvimento sustentável aliado à redução das emissões de carbono.


Este projeto analisa a evolução temporal das emissões de dióxido de carbono (CO₂) provenientes do setor industrial, em escala global e nacional. Por meio da aplicação de modelos analíticos de séries temporais, busca-se compreender como a atividade industrial impacta o balanço de carbono e avaliar se avanços tecnológicos e mudanças na matriz energética têm contribuído efetivamente para a descarbonização da produção. Além disso, o estudo pretende projetar cenários futuros, auxiliando na tomada de decisão estratégica.


A base de dados utilizada será proveniente do Our World in Data, com dados compilados do Global Carbon Project. O conjunto contém séries históricas de emissões de CO₂ (em milhões de toneladas), com granularidade anual, abrangendo diferentes países e períodos históricos extensos. Entre as principais variáveis estão: entidade (país/região), ano, emissões totais de CO₂ e emissões específicas do setor industrial. Os dados estão estruturados em formato .csv no padrão “tidy data”, permitindo fácil manipulação em ferramentas como Python ou R.


A motivação do projeto está relacionada a três fatores principais: a complexidade da descarbonização do setor industrial, considerado um dos mais difíceis de adaptar (hard-to-abate); a crescente pressão de investidores e consumidores por transparência ambiental, associada a critérios ESG; e a necessidade de avaliar, com base em dados, se a adoção de tecnologias mais limpas e fontes renováveis está gerando impactos reais ao longo do tempo.


A análise se justifica pela natureza temporal dos dados de emissões, que apresentam tendências e possíveis padrões associados a ciclos econômicos e avanços tecnológicos. A abordagem de séries temporais permite identificar pontos de inflexão — momentos em que as emissões começam a desacelerar ou reduzir mesmo diante do crescimento produtivo — e avaliar a viabilidade estatística de metas globais, como o “Net Zero”. Assim, o objetivo do projeto é desenvolver um produto analítico capaz de modelar, prever e comparar as emissões industriais de CO₂, gerando diagnósticos sobre a velocidade da descarbonização e projeções para os próximos anos.


De acordo com o Our World in Data, as emissões de gases de efeito estufa são o principal fator das mudanças climáticas atuais, sendo resultado direto de atividades como a queima de combustíveis fósseis e a produção industrial. Nesse sentido, a transformação dos sistemas energéticos e industriais, aliada a avanços tecnológicos e políticas públicas eficazes, é essencial para reduzir emissões sem comprometer o desenvolvimento econômico e social.



# **Referencial Teórico**

A análise de emissões de CO₂ por meio de séries temporais tem sido
amplamente explorada na literatura como ferramenta para compreensão
de tendências históricas e projeções futuras no contexto das mudanças
climáticas. Estudos como os de Ritchie, Roser e Rosado (2023) destacam
a importância de dados históricos consolidados para avaliar o impacto das
atividades humanas no aumento das emissões globais, permitindo
análises comparativas entre países e setores ao longo do tempo.


Diversas abordagens metodológicas têm sido utilizadas para modelagem
de séries temporais ambientais. Modelos estatísticos clássicos, como o
ARIMA (AutoRegressive Integrated Moving Average), são frequentemente
aplicados devido à sua capacidade de capturar tendências e padrões
temporais em dados históricos (BOX; JENKINS; REINSEL, 2015). Esses
modelos apresentam como principais vantagens a interpretabilidade e a
eficiência computacional, sendo amplamente utilizados em contextos com
dados estruturados. No entanto, possuem limitações na modelagem de
padrões não lineares complexos e podem apresentar desempenho
reduzido em séries com alta variabilidade.


Alternativamente, métodos baseados em aprendizado de máquina, como
Redes Neurais Recorrentes e modelos LSTM (Long Short-Term Memory),
têm sido utilizados para previsão de séries temporais, especialmente em
contextos com grande volume de dados e comportamentos não lineares
(HOCHREITER; SCHMIDHUBER, 1997). Essas abordagens permitem
capturar dependências de longo prazo e relações complexas entre
variáveis, mas apresentam desvantagens como maior complexidade
computacional, necessidade de grandes volumes de dados para
treinamento e menor interpretabilidade dos resultados (GOODFELLOW;
BENGIO; COURVILLE, 2016). No contexto específico de emissões
industriais, trabalhos recentes indicam que fatores como crescimento
econômico, matriz energética e políticas públicas influenciam diretamente os padrões observados (GLOBAL CARBON PROJECT, 2023). Além
disso, a literatura aponta a existência de possíveis pontos de inflexão
associados à transição energética, nos quais o crescimento econômico
passa a ocorrer com menor intensidade de emissões, fenômeno
frequentemente discutido no contexto da chamada curva de Kuznets
ambiental (STERN, 2004).


Os principais conceitos envolvidos neste projeto incluem:

• Séries temporais e sua decomposição em tendência, sazonalidade
e ruído (HAMILTON, 1994);

• Modelos preditivos, incluindo ARIMA, regressão e modelos de
machine learning;

• Avaliação de modelos por meio de métricas como erro médio
absoluto (MAE) e raiz do erro quadrático médio (RMSE) (HYNDMAN;
ATHANASOPOULOS, 2021);

• Análise de tendências e identificação de pontos de inflexão;

• Descarbonização e indicadores ambientais relacionados a critérios
ESG.


Assim, o presente trabalho se insere na interseção entre análise
estatística e sustentabilidade, buscando aplicar métodos consolidados da
literatura para avaliar a evolução e as perspectivas das emissões
industriais, contribuindo para o entendimento dos desafios associados à
transição para uma economia de baixo carbono.
